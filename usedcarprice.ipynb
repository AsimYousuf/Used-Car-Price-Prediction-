```python
import pandas as pd
```

```python
df = pd.read_csv('usedcarprice.csv')
```

```python
df
```

```python
df2 = df.copy()       #using a copy for smoother operations
```

```python
df['company'] = df2['name'].str.split(' ').str.get(0)
```

```python
df
```

```python
df = df.drop(['name'], axis =1) #we will use company instead of name to increase efficiency of operation
```

```python
df
```

```python
df.info()
```

```python
df = pd.concat([df.drop('company', axis = 1), pd.get_dummies(df.company).add_prefix('company_')], axis =1)
#creating dummies for company
```

```python
df
```

```python
df = df.drop(['color'], axis =1) #dropping color as it is too subjective to be statistically significant
```

```python
df['miles'] = df['miles'].str.replace('miles','')
df['miles'] = df['miles'].str.replace(',','')
#making miles ready to be converted to int type
```

```python
df.miles
```

```python
df['miles'] = df['miles'].astype(int)    #changing miles to integer type
```

```python
df.miles
```

```python
df['price'] = df['price'].str.replace('$','')
df['price'] = df['price'].str.replace(',','')
df['price'] = df['price'].astype(int) 
```

```python
df.price
```

```python
df['accidents'] = df2['condition'].str.split(', ').str.get(0)
df['prev_owners'] = df2['condition'].str.split(', ').str.get(1)
df = df.drop(['condition'], axis =1)
```

```python
df
```

```python
df['prev_owners'] = df['prev_owners'].str.split(' ').str.get(0)
df['prev_owners'] = df['prev_owners'].astype(int) 
```

```python
df.prev_owners
```

```python
df['accidents'] = df['accidents'].str.split(' ').str.get(0)
df['accidents'] = df['accidents'].str.replace('No','0')
df['accidents'] = df['accidents'].astype(int)
```

```python
df.info()
```

```python
# year, miles, rolls royce, and prev owners are statistically significant with respect to price 
#hence will will build a linear regression model using these two variables
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictors = ['year', 'miles','company_Rolls-Royce','prev_owners']
target = 'price'
```

```python
train =df[df['year']<2018].copy()
test = df[df['year']>=2018].copy()
```

```python
reg.fit(train[predictors], train[target])
```

```python
predictions = reg.predict(test[predictors])
```

```python
test["predictions"] = predictions
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test['price'],test['predictions'])
```
